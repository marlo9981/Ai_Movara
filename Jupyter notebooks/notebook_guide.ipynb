# Understanding the .ipynb File Structure

A .ipynb file is a JSON document that contains:
- **Cells**: Code, markdown, or raw cells
- **Metadata**: Information about the notebook, kernel, etc.
- **nbformat**: Version information

You can load and inspect a .ipynb file using Python:

````python
import json

# Load the current notebook
with open('notebook_guide.ipynb', 'r') as f:
    nb = json.load(f)

print("Notebook structure keys:", list(nb.keys()))
print("Number of cells:", len(nb['cells']))
print("Metadata:", nb['metadata'])
````

# Adding Code Cells

Code cells contain executable Python code. You can run them to see output below the cell.

````python
# Example code cell
print("Hello from a code cell!")
x = 5
y = 10
print(f"x + y = {x + y}")
````

# Including Metadata

Metadata provides information about the notebook and cells. Common metadata includes:
- Kernel specification
- Language info
- Cell execution count
- Custom tags

````python
import json

# Load and display metadata
with open('notebook_guide.ipynb', 'r') as f:
    nb = json.load(f)

print("Notebook metadata:")
print(json.dumps(nb['metadata'], indent=2))

# Cell metadata (this cell has none by default)
print("\nCell metadata for this cell:")
print("None (can be added via cell toolbar or programmatically)")
````
